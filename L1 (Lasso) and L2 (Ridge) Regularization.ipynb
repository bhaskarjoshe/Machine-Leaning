```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso , Ridge

warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('C:\\Data\\Melbourne_housing_FULL.csv')
df.head()
```

```python
df.nunique()
```

```python
col_to_use=df.drop(['Address','Date','YearBuilt','Longtitude','Lattitude','Regionname'],axis=1)
new_df=pd.DataFrame(col_to_use)
new_df.head()
```

```python
new_df.isna().sum()
```

```python
cols_to_fill_zero=['Propertycount','Car','Bedroom2','Bathroom','Distance']
new_df[cols_to_fill_zero]=new_df[cols_to_fill_zero].fillna(0)
new_df.isna().sum()
```

```python
new_df['Landsize']=new_df['Landsize'].fillna(new_df.Landsize.mean())
new_df['BuildingArea']=new_df['BuildingArea'].fillna(new_df.BuildingArea.mean())
new_df.isna().sum()
```

```python
new_df.dropna(inplace=True)
new_df.isna().sum()
```

```python
new_df=pd.get_dummies(new_df,drop_first=True)
new_df
```

```python
x=new_df.drop('Price',axis=1)
y=new_df['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
model=LinearRegression().fit(x_train,y_train)
```

```python
print('Model Score (Test Dataset): {} \nModel Score (Train Dataset): {}'.format(model.score(x_test,y_test),model.score(x_train,y_train)))
```

## Lasso Regression (L1 Regularization)

```python
lasso_reg_model=Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg_model.fit(x_train,y_train)
```

```python
print('Lasso-\nModel Score (Test Dataset): {} \nModel Score (Train Dataset): {}'.format(lasso_reg_model.score(x_test,y_test),lasso_reg_model.score(x_train,y_train)))
```

## Ridge Regression (L2 Regularization)

```python
ridge_reg_model=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg_model.fit(x_train,y_train)
```

```python
print('Ridge-\nModel Score (Test Dataset): {} \nModel Score (Train Dataset): {}'.format(ridge_reg_model.score(x_test,y_test),ridge_reg_model.score(x_train,y_train)))
```
